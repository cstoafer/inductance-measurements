### Get Data

The lecroy scope has software that was used to download output csv files from the scope readings.  These were terribly formatted csv's with all the data in one column and incomplete information.  The following code parses these csv files for the two signals.  The voltage scale is determined from the csv file, but the time scale is not.  The time scale does not matter because we only care about amplitude and phase information.

```python
def get_data(folder, freq):
    '''
    Extracts data from csv from the small Lecroy scope
    '''
    with open('scope data/{0}/{1}hz.csv'.format(folder, freq), 'rb') as csvfile:
        csv_file_object = csv.reader(csvfile)
        #num_points = np.int(csv_file_object.next())
        data = []
        for row in csv_file_object:
            data.extend(row)

    num_points = np.int(data[0])

    # Lecroy saves the voltage multiplier for each channel in a strange way.  The next 4 lines takes care of that.
    ch1_gain = data[num_points+6]
    ch1_gain = np.float(ch1_gain.split('V')[0]) if re.search("V", ch1_gain) else 1e-3*np.float(ch1_gain)
    ch2_gain = data[2*num_points+47]
    ch2_gain = np.float(ch2_gain.split('V')[0]) if re.search("V", ch2_gain) else 1e-3*np.float(ch2_gain)
    #print ch1_gain, ch2_gain

    signals = np.empty([num_points-2,3])
    signals[:,0] = ch1_gain / 50 * np.array(data[5:num_points+3], dtype=float)
    signals[:,1] = ch2_gain / 50 * np.array(data[num_points+47:2*num_points+45], dtype=float)

    ch1_avg = np.mean(signals[:,0])
    ch2_avg = np.mean(signals[:,1])
    signals[:,0] -= ch1_avg
    signals[:,1] -= ch2_avg
    
    # Unfortunatley, the output from the Lecroy scope has the wrong time/div information, so we will have to determine this by
    # inferring it from the frequency.  This has been confirmed with readings directly on the scope

    # Lecroy has terrible output files.  The time/div is not labelled with s, ms, or us, so we have to figure it 
    # out using the frequency
    #time_div = data[num_points+4]
    # remove non-numeric
    #non_decimal = re.compile(r'[^\d.]+')
    #time_div = non_decimal.sub('', time_div)
    #time_div = np.float(time_div)
    #while freq * time_div > 1:
    #    time_div *= 1e-3
    #signals[:,0] = np.arange(num_points-2)
    #signals[:,0] *= time_div

    #print signals
    #df = pd.DataFrame(signals, columns=['time (ms)','Ch1','Ch2'])
    
    return signals
```

### Fit sine wave to signals

A sine wave is fit to each signal in order to determine signal amplitude and phase.  The initial conditions for fitting a sine very are very important for the fit to converge because it is a non-linear fit.  It is difficult to fit four parameters at once (frequency, amplitude, phase, and offset), so the frequency and phase are fit first, then all the of the parameters are fit with two separate fit calls.  

Initial conditions are imperative for a good fit and the requirements change depending on how the signal looks from the scope.  For example, the sine signal will have different intial phases.  Therefore, information about the signal that is being fit is used to select the initial conditions.  Phase is chosen based on the signal value and its derivative towards the beginning of the signal.  Frequency is chosen based on a peak in a FFT of the signal.  Amplitude is the maximum value of the signal.

The error of the fit is checked and a warning output is generated if a fit does not converge.

```python
from scipy.optimize import leastsq
import scipy.fftpack as fftpack
def fit_sine(signal):
    guess_A = np.max(signal)
    if np.abs(signal[0]) < guess_A/2:
        guess_ph = -(signal[20] - signal[0]) / np.abs(signal[20] - signal[0]) + 1
    else:
        guess_ph = -signal[0]/np.abs(signal[0]) + 2
    initial_ph = guess_ph
    yhat = fftpack.rfft(signals[:,0])
    idx = (yhat**2).argmax()
    freqs = fftpack.rfftfreq(len(signal), d = 1)
    guess_freq = freqs[idx]
    initial_freq = guess_freq

    t = np.arange(len(signal))
    error = 1000
    flag=0
    while error > 300 and flag<=1:
        # Define the function to optimize, in this case, we want to minimize the difference
        # between the actual data and our "guessed" parameters
        optimize_func = lambda x: guess_A*np.sin(x[0]*2*np.pi*t+x[1]) - signal
        est_freq, est_ph = leastsq(optimize_func, [guess_freq, guess_ph])[0]

        error = np.sum(np.abs(guess_A*np.sin(est_freq*2*np.pi*t+est_ph) - signal))

        # one more time with all 4 parameters
        guess_freq = est_freq
        guess_ph = est_ph
        guess_offset = 1
        optimize_func = lambda x: x[0]*np.sin(x[1]*2*np.pi*t+x[2]) + x[3] - signal
        est_A, est_freq, est_ph, est_offset = leastsq(optimize_func, [guess_A, guess_freq, guess_ph, guess_offset])[0]

        # Some of the sine signals would not converge with the initial phase settings
        # By reviewing the initial conditions, I found that if the phase was chosen to be opposite (negative) of the
        # chosen phase, then the fit would converge.
        error = np.sum(np.abs(est_A*np.sin(est_freq*2*np.pi*t+est_ph) + est_offset - signal))
        if error > 300:
            guess_ph = -initial_ph/np.abs(initial_ph)
            guess_freq = initial_freq
            guess_A = np.max(signal)
            guess_offset = 1
            flag+=1
    
    if est_ph < -np.pi:
        est_ph += 2*np.pi
    if est_ph > np.pi:
        est_ph -= 2*np.pi
    return ([est_A, est_freq, est_ph, est_offset], error)
```

### Determine the inductance

We determine the inductance by using the amplitude of the voltage across the resistor and the voltage across the circuit.  The following shows how we determine the inductance from this information.  $|V_C|$ and $|V_R|$ are the amplitudes of the voltage across the circuit and resistor, respectively.  $R_{meas}$ is the resistor we measure across, $R_{ind}$ is the DC resistance of the inductor, $\omega$ is related to the frequency, and $I$ is the current in the circuit. 
$$ V_C = (i\omega L + R_{meas} + R_{inductor})I $$
$$ V_R = R_{meas}I $$

$$ |V_C| = \sqrt{V_CV_C^*} = |I|\sqrt{\omega^2L^2 + (R_{meas} + R_{ind})^2} $$
$$ |V_R| = R_{meas} |I| $$

By rearranging the terms in the equation for $|V_C|$, then using the equation $|I|=|V_R|/R_{meas}$ to plug-in for $|I|$:

$$ \omega L = \sqrt{(|V_C| / |I|)^2 - (R_{meas} + R_{ind})^2} = \sqrt{R_{meas}^2(|V_C| / |V_R|)^2 - (R_{meas} + R_{ind})^2} $$

$$ \omega L = R_{meas} \sqrt{\left(\frac{|V_C|}{|V_R|}\right)^2 - \left(1 + \frac{R_{ind}}{R_{meas}}\right)^2} $$

$$\textrm{or}$$

$$ L = \frac{R_{meas}}{2\pi f} \sqrt{\left(\frac{|V_C|}{|V_R|}\right)^2 - \left(1 + \frac{R_{ind}}{R_{meas}}\right)^2} $$

```python
def determine_inductance(folder, freq, R_meas, R_ind):
    '''
    Used if a scope is used to measure across a resistor in series with an inductor, where the inductor also has a significant resistance.
    '''
    freqs = np.asarray(freq)
    V_circ = [data[folder][freq]['circuit'][0] for freq in freqs]
    V_r = [data[folder][freq]['resistor'][0] for freq in freqs]
    V_circ = np.asarray(V_circ)
    V_r = np.asarray(V_r)
    #inductance = 1/(2*np.pi*freqs)*R_meas*np.sqrt(V_circ**2 / V_r**2 - 1)
    inductance = 1/(2*np.pi*freqs)*np.sqrt((R_meas * V_circ / V_r)**2 - (R_meas + R_ind)**2 )
    
    return inductance if freq.ndim > 0 else inductance
```

```python
for folder in folders:
    if '3ohm' in folder:
        R_ind = 2.1 if 'stainless' in folder else 2.4
    else:
        R_ind=0
    #R_ind=1.5
    #V_circ = [data[folder][freq]['circuit'][0] for freq in freqs]
    #V_r = [data[folder][freq]['resistor'][0] for freq in freqs]
    R_meas = 50 if '50ohm' in folder else 2.7
    inductance = determine_inductance(folder, freqs, R_meas, R_ind)
    #print inductance
    for i, freq in enumerate(freqs):
        data[folder][freq]['inductance'] = inductance[i]
        data[folder][freq]['mu'] = inductance[i] / data[folders[0]][freq]['inductance']
```

The terms in the square root are the most important in determining $\omega L$.  We will take a look at them separately. We use this value as measured.  $R_{ind}$ is due to the wire windings around our sample.  We can measure the value with a DC voltmeter, as we have done, but we can also use phase shift information from our voltage measurements of the circuit and measurement resistor to try to reduce error in measuring the resistance and other experimental effects.  We will come back to this later.

### Analyzing measurement sensitivity

The analysis above shows that using different measurement resistors changes the sensitivity range of frequencies.  For our samples and desired frequency range, we are limited in lower frequencies (other cases lead to different limitations in higher frequencies).  This becomes obvious in reviewing the equations that determine our voltage measurements: the circuit voltage, $|V_C| = |I|\sqrt{\omega^2L^2 + (R_{meas} + R_{ind})^2}$, and the measurement resistor voltage, $|V_R| = R_{meas} |I|$.  When we take the ratio of these equations, which is what we use to determine %\omega L$, we have:

$$ \frac{|V_C|}{|V_R|} = \frac{1}{R_{meas}}\sqrt{(\omega L)^2 + (R_{meas} + R_{ind})^2} $$

Looking at the terms in the square root, $(\omega L)^2$ and $(R_{meas} + R_{ind})^2$, we do not want the second term to dominate over the first.  In that case the measurement would be insensitive to changes in frequency.  Therefore we want $\omega L \approx R_{meas} + R_{ind}$.  Since we know we have a lower bound in frequency ($\omega$ or $f$), we can setup a relation such that $\omega L$ is at least a certain proportion to $(R_{meas} + R_{ind})$.  Let's set the cutoff such that $\omega L$ is greater than $1/100^{th}$ of $(R_{meas} + R_{ind})$.

$$ \omega L \gtrsim 1/100(R_{meas} + R_{ind}) $$

Let's look at the case where $R_{meas}$ is 2.7 ohms.  We also know that $R_{ind}$ is approximately 2.3 ohms by direct measurement.  The inductance $L$ of stainless steel is about $0.01$ mH, which is $10^{-5}$ H and the inductance of the HiperCo sample is around 10 times that for moderate applied magnetic field, so is $10^{-4}$ H.  Also using the definition $\omega = 2\pi f$.

$$ 2\pi fL \gtrsim 1/100(2.7 + 2.3) $$

$$ f \gtrsim \frac{5}{200\pi L} $$
<br>
<center>For the stainless sample, using $L \approx 10^{-5}$ H</center>

$$ f \gtrsim 800~ \textrm{Hz} $$
<br>
<center>For the HiperCo sample, using $L \approx 10^{-4}$ H</center>

$$ f \gtrsim 80~ \textrm{Hz} $$

This validates our observation about a lower limitation in being able to measure the voltage amplitude ratio with our experimental parameters.  If we want to get reliable measurements below these frequencies, we need to reduce $R_{meas}$, but due to other limitations can only reduce it to $\sim 1$ ohm.
```python
def determine_inductance_phase(folder, freq, R_meas):
    '''
    Used if a scope is used to measure across a resistor in series with an inductor, where the inductor also has a significant resistance. 
    The resistance with the resistor is determined using phase infomration of the signal.
    '''
    freqs = np.asarray(freq)
    V_circ = [data[folder][freq]['circuit'][0] for freq in freqs]
    V_r = [data[folder][freq]['resistor'][0] for freq in freqs]
    phase = [data[folder][freq]['circuit'][2] - data[folder][freq]['resistor'][2] for freq in freqs]
    V_circ = np.asarray(V_circ)
    V_r = np.asarray(V_r)
    phase = np.asarray(phase)
    #inductance = 1/(2*np.pi*freqs)*R_meas*np.sqrt(V_circ**2 / V_r**2 - 1)
    inductance = R_meas/(2*np.pi*freqs) * V_circ / V_r * np.tan(phase) / np.sqrt(1 + (np.tan(phase))**2)
    
    return inductance if freq.ndim > 0 else inductance
```

```python
folders = [folder for folder in folders if '3ohm' in folder]
for folder in folders:
    R_meas = 2.7
    inductance = determine_inductance_phase(folder, freqs, R_meas)
    print inductance
    for i, freq in enumerate(freqs):
        data[folder][freq]['inductance_phase'] = inductance[i]
        data[folder][freq]['mu_phase'] = inductance[i] / 0.0223
```
