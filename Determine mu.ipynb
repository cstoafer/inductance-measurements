# Determining relative permeability, $\mu$, of ferromagnetic material used in the ferritic walls of HBT-EP

I perform my graduate research with a plasma physics and nuclear fusion group on a device called a tokamak; our particular device is call The High-Beta Tokamak - Extended Pulse (HBT-EP).  The group is interested in studying the effects of having a ferritic material (a material that can be magnetized, such as iron) in a wall near the plasma edge.  This is because future nuclear fusion reactors require materials with low neutron activation (materials that will not quickly deteriorate when constantly being hit by neutrons).  These types of materials are typically ferritic.  Tokamaks use magnetic field to confine and stabilize plasmas and these field will change with the presence of ferritic materials.  Our goal is to study these effects, especially on plasma confinement and stability, which are crucial to nuclear fusion energy.

A material was chosen for the ferritic wall on HBT-EP to provide a large effect with a limited thickness of the material.  The relative permeability, $\mu$, of the material chosen, HiperCo 50, which is a mix of $\sim 49\%$ cobalt, $\sim 49\%$ iron, $\sim 2\%$ vanadium, with other trace elements, including nickel.  According to manufacturer specifications, the material should have a value $\mu = 8$ with an applied field of $0.35$ Tesla, the nominal toroidal field in HBT-EP. 

Although this is not my area of expertise, I volunteered to help the group by performing careful measurements and data analysis.  The results were initially unexpected, requiring a more subtle understanding and description.  This document outlines the measurement and analysis performed to test the $\mu$ value of the material in the order the experiments were performed.  It outlines our learning process through this endeavour.

### Description of Experiment

A sample of the HiperCo 50 material was machined to make a rectangular slab 10 cm long, 1.05cm wide, and 0.55 cm thick.  A thin wire was then wrapped around the sample for a total of 171 turns over 9.5 cm.  A similar sample of stainless steel (316 steel, non-ferritic material) was also made for comparison.  With the wire wrapped around the samples, these are called solenoids.  The inductance of each solenoid depends on the relative permeability, $\mu$, of the core sample.  This is described in an <a href="http://hyperphysics.phy-astr.gsu.edu/hbase/electric/indsol.html">explanation</a> by HyperPhysics, which is a great source for learning about physical phenomena.  The diagram from the HyperPhysics explanation is shown below.

<img src="figures/sol3.jpg" width="500" />

Based on the formula from HyperPhysics, $L = \mu N^2A/l$, since we know $N=171$ (number of turns), $A = 0.5775$cm$^2$ (cross-sectional area), and $l = 9.5$cm (length of windings), by measuring the inductance of the sample, $L$, we will be able to determine $\mu$.  To do this, we connect the sample as an inductor to a simple circuit to determine the inductance.

The $\mu$ value of the material also depends on the external magnetic field, $B_{external}$ outside of the sample.  This is a separate magnetic field than the one created by the current, $I$, through the coil.  $B_{external}$ is shown with the green arrows below, whereas $B_{applied}$ is due to the current in the coil as the light blue lines.  My lab is interested in measureing $\mu$ at different magnetic fields because the material we use is placed in an environment with a large external magnetic field (~0.33 Tesla or 3,300 Gauss).  With magnetic material, such as HiperCo 50, there are small domains within the material that are magnetized in a certain direction.  With no external magnetic field, these domain orientations are randomly distributed.  The domains then align with the applied magnetic field from the current through the coil, $B_{applied}$.  When there is an external magnetic field, the domains are preferentially aligned to $B_{external}$ which decreases how many domains can align with $B_{applied}$ and how well they can align with it.  Therefore, we expect the effective $\mu$ and inductance to decrease with higher $B_{external}$.

<img src="figures/sol3_Bext_perp.jpg" width="500" />

The two leads of the windings of the HiperCo 50 sample were connected to a circuit for the measurement, as seen in the figure below.  A function generator was used to drive sine waves at various frequencies, which was hooked up across the test sample ($L_{sample}$ and $R_{sample}$) and a 2.7$\Omega$ resistor ($R_{meas}$) in series.  A scope was used to measure the time traces of both the voltage across the circuit (including the test sample and $R_{meas}$) as well as the voltage across $R_{meas}$.

<img src="figures/circuit.jpg" width="700" />

The test sample has an inductance, $L_{sample}$, which we are trying to determine.  The sample also has a non-negligible resistance, $R_{sample}$ because of the length of wire wrapped around the sample.  The measurement resistor, $R_{meas}$, is used to determine the current in the circuit as a function of time because voltage and current have a direct relation across a resistor, $I = V_{R}/R_{meas}$.  We will see later exactly how these voltage measurements are used to determine the inductance of the sample.

Three measurements (1, 3, and 7) were taken at frequencies in every decade from $10^2$ Hz to $10^5$ Hz (i.e. at 100 Hz, 300 Hz, 700Hz, 1000 Hz, 3000 Hz, 7000 Hz ... ).  This was done with both the HiperCo 50 and stainless steel samples.  These measurements were made on the HiperCo 50 sample with various external magnetic field induced by a large electromagnet.  This allows us to test the saturation field and mu at various applied magnetic fields.

### Get Data

The lecroy scope has software that was used to download output csv files from the scope readings.  These were terribly formatted csv's with all the data in one column and incomplete information.  The following code parses these csv files for the two signals.  The voltage scale is determined from the csv file, but the time scale is not.  The time scale does not matter because we only care about amplitude and phase information.

```python
def get_data(folder, freq):
    '''
    Extracts data from csv from the small Lecroy scope
    '''
    with open('scope data/{0}/{1}hz.csv'.format(folder, freq), 'rb') as csvfile:
        csv_file_object = csv.reader(csvfile)
        #num_points = np.int(csv_file_object.next())
        data = []
        for row in csv_file_object:
            data.extend(row)

    num_points = np.int(data[0])

    # Lecroy saves the voltage multiplier for each channel in a strange way.  The next 4 lines takes care of that.
    ch1_gain = data[num_points+6]
    ch1_gain = np.float(ch1_gain.split('V')[0]) if re.search("V", ch1_gain) else 1e-3*np.float(ch1_gain)
    ch2_gain = data[2*num_points+47]
    ch2_gain = np.float(ch2_gain.split('V')[0]) if re.search("V", ch2_gain) else 1e-3*np.float(ch2_gain)
    #print ch1_gain, ch2_gain

    signals = np.empty([num_points-2,3])
    signals[:,0] = ch1_gain / 50 * np.array(data[5:num_points+3], dtype=float)
    signals[:,1] = ch2_gain / 50 * np.array(data[num_points+47:2*num_points+45], dtype=float)

    ch1_avg = np.mean(signals[:,0])
    ch2_avg = np.mean(signals[:,1])
    signals[:,0] -= ch1_avg
    signals[:,1] -= ch2_avg
    
    # Unfortunatley, the output from the Lecroy scope has the wrong time/div information, so we will have to determine this by
    # inferring it from the frequency.  This has been confirmed with readings directly on the scope

    # Lecroy has terrible output files.  The time/div is not labelled with s, ms, or us, so we have to figure it 
    # out using the frequency
    #time_div = data[num_points+4]
    # remove non-numeric
    #non_decimal = re.compile(r'[^\d.]+')
    #time_div = non_decimal.sub('', time_div)
    #time_div = np.float(time_div)
    #while freq * time_div > 1:
    #    time_div *= 1e-3
    #signals[:,0] = np.arange(num_points-2)
    #signals[:,0] *= time_div

    #print signals
    #df = pd.DataFrame(signals, columns=['time (ms)','Ch1','Ch2'])
    
    return signals
```

### Fit sine wave to signals

A sine wave is fit to each signal in order to determine signal amplitude and phase.  The initial conditions for fitting a sine very are very important for the fit to converge because it is a non-linear fit.  It is difficult to fit four parameters at once (frequency, amplitude, phase, and offset), so the frequency and phase are fit first, then all the of the parameters are fit with two separate fit calls.  

Initial conditions are imperative for a good fit and the requirements change depending on how the signal looks from the scope.  For example, the sine signal will have different intial phases.  Therefore, information about the signal that is being fit is used to select the initial conditions.  Phase is chosen based on the signal value and its derivative towards the beginning of the signal.  Frequency is chosen based on a peak in a FFT of the signal.  Amplitude is the maximum value of the signal.

The error of the fit is checked and a warning output is generated if a fit does not converge.

```python
from scipy.optimize import leastsq
import scipy.fftpack as fftpack
def fit_sine(signal):
    guess_A = np.max(signal)
    if np.abs(signal[0]) < guess_A/2:
        guess_ph = -(signal[20] - signal[0]) / np.abs(signal[20] - signal[0]) + 1
    else:
        guess_ph = -signal[0]/np.abs(signal[0]) + 2
    initial_ph = guess_ph
    yhat = fftpack.rfft(signals[:,0])
    idx = (yhat**2).argmax()
    freqs = fftpack.rfftfreq(len(signal), d = 1)
    guess_freq = freqs[idx]
    initial_freq = guess_freq

    t = np.arange(len(signal))
    error = 1000
    flag=0
    while error > 300 and flag<=1:
        # Define the function to optimize, in this case, we want to minimize the difference
        # between the actual data and our "guessed" parameters
        optimize_func = lambda x: guess_A*np.sin(x[0]*2*np.pi*t+x[1]) - signal
        est_freq, est_ph = leastsq(optimize_func, [guess_freq, guess_ph])[0]

        error = np.sum(np.abs(guess_A*np.sin(est_freq*2*np.pi*t+est_ph) - signal))

        # one more time with all 4 parameters
        guess_freq = est_freq
        guess_ph = est_ph
        guess_offset = 1
        optimize_func = lambda x: x[0]*np.sin(x[1]*2*np.pi*t+x[2]) + x[3] - signal
        est_A, est_freq, est_ph, est_offset = leastsq(optimize_func, [guess_A, guess_freq, guess_ph, guess_offset])[0]

        # Some of the sine signals would not converge with the initial phase settings
        # By reviewing the initial conditions, I found that if the phase was chosen to be opposite (negative) of the
        # chosen phase, then the fit would converge.
        error = np.sum(np.abs(est_A*np.sin(est_freq*2*np.pi*t+est_ph) + est_offset - signal))
        if error > 300:
            guess_ph = -initial_ph/np.abs(initial_ph)
            guess_freq = initial_freq
            guess_A = np.max(signal)
            guess_offset = 1
            flag+=1
    
    if est_ph < -np.pi:
        est_ph += 2*np.pi
    if est_ph > np.pi:
        est_ph -= 2*np.pi
    return ([est_A, est_freq, est_ph, est_offset], error)
```

### Determine the inductance

We determine the inductance by using the amplitude of the voltage across the resistor and the voltage across the circuit.  The following shows how we determine the inductance from this information.  $|V_C|$ and $|V_R|$ are the amplitudes of the voltage across the circuit and resistor, respectively.  $R_{meas}$ is the resistor we measure across, $R_{ind}$ is the DC resistance of the inductor, $\omega$ is related to the frequency, and $I$ is the current in the circuit. 
$$ V_C = (i\omega L + R_{meas} + R_{inductor})I $$
$$ V_R = R_{meas}I $$

$$ |V_C| = \sqrt{V_CV_C^*} = |I|\sqrt{\omega^2L^2 + (R_{meas} + R_{ind})^2} $$
$$ |V_R| = R_{meas} |I| $$

By rearranging the terms in the equation for $|V_C|$, then using the equation $|I|=|V_R|/R_{meas}$ to plug-in for $|I|$:

$$ \omega L = \sqrt{(|V_C| / |I|)^2 - (R_{meas} + R_{ind})^2} = \sqrt{R_{meas}^2(|V_C| / |V_R|)^2 - (R_{meas} + R_{ind})^2} $$

$$ \omega L = R_{meas} \sqrt{\left(\frac{|V_C|}{|V_R|}\right)^2 - \left(1 + \frac{R_{ind}}{R_{meas}}\right)^2} $$

$$\textrm{or}$$

$$ L = \frac{R_{meas}}{2\pi f} \sqrt{\left(\frac{|V_C|}{|V_R|}\right)^2 - \left(1 + \frac{R_{ind}}{R_{meas}}\right)^2} $$

```python
def determine_inductance(folder, freq, R_meas, R_ind):
    '''
    Used if a scope is used to measure across a resistor in series with an inductor, where the inductor also has a significant resistance.
    '''
    freqs = np.asarray(freq)
    V_circ = [data[folder][freq]['circuit']['amp'] for freq in freqs]
    V_r = [data[folder][freq]['resistor']['amp'] for freq in freqs]
    V_circ = np.asarray(V_circ)
    V_r = np.asarray(V_r)
    #inductance = 1/(2*np.pi*freqs)*R_meas*np.sqrt(V_circ**2 / V_r**2 - 1)
    inductance = 1/(2*np.pi*freqs)*np.sqrt((R_meas * V_circ / V_r)**2 - (R_meas + R_ind)**2 )
    
    return inductance if freq.ndim > 0 else inductance
```

```python
for folder in folders:
    if '3ohm' in folder:
        R_ind = 2.1 if 'stainless' in folder else 2.4
    else:
        R_ind=0
    #R_ind=1.5
    #V_circ = [data[folder][freq]['circuit'][0] for freq in freqs]
    #V_r = [data[folder][freq]['resistor'][0] for freq in freqs]
    R_meas = 50 if '50ohm' in folder else 2.7
    inductance = determine_inductance(folder, freqs, R_meas, R_ind)
    #print inductance
    for i, freq in enumerate(freqs):
        data[folder][freq]['inductance'] = inductance[i]
```

The constant values of $\omega L$ at lower frequencies implies a limitation in what we can measure because the inductance should be close to constant at lower frequencies and $\omega$ should be increasing with frequency by definition.  Let's review the equation we used to determine $\omega L$:

$$ \omega L = R_{meas} \sqrt{\left(\frac{|V_C|}{|V_R|}\right)^2 - \left(1 + \frac{R_{ind}}{R_{meas}}\right)^2} $$

The terms in the square root are the most important in determining $\omega L$.  We will take a look at them separately.  The first term $(|V_C|/|V_R|)^2$ is what we actually measure (the ratio of the amplitudes of the voltage signals across the circuit and the measurement resistor).  The second term, $(1 + R_{ind}/R_{meas})^2$ affects the measurement in the more subtle way.  $R_{meas}$ has a well known value because it has a nominal value by the manufacturer and we can directly measure it (and confirm it is close to the nominal value).  We use this value as measured.  $R_{ind}$ is due to the wire windings around our sample.  We can measure the value with a DC voltmeter, as we have done, but we can also use phase shift information from our voltage measurements of the circuit and measurement resistor to try to reduce error in measuring the resistance and other experimental effects.  We will come back to this later.

### Analyzing measurement sensitivity

The analysis above shows that using different measurement resistors changes the sensitivity range of frequencies.  For our samples and desired frequency range, we are limited in lower frequencies (other cases lead to different limitations in higher frequencies).  This becomes obvious in reviewing the equations that determine our voltage measurements: the circuit voltage, $|V_C| = |I|\sqrt{\omega^2L^2 + (R_{meas} + R_{ind})^2}$, and the measurement resistor voltage, $|V_R| = R_{meas} |I|$.  When we take the ratio of these equations, which is what we use to determine %\omega L$, we have:

$$ \frac{|V_C|}{|V_R|} = \frac{1}{R_{meas}}\sqrt{(\omega L)^2 + (R_{meas} + R_{ind})^2} $$

Looking at the terms in the square root, $(\omega L)^2$ and $(R_{meas} + R_{ind})^2$, we do not want the second term to dominate over the first.  In that case the measurement would be insensitive to changes in frequency.  Therefore we want $\omega L \approx R_{meas} + R_{ind}$.  Since we know we have a lower bound in frequency ($\omega$ or $f$), we can setup a relation such that $\omega L$ is at least a certain proportion to $(R_{meas} + R_{ind})$.  Let's set the cutoff such that $\omega L$ is greater than $1/100^{th}$ of $(R_{meas} + R_{ind})$.

$$ \omega L \gtrsim 1/100(R_{meas} + R_{ind}) $$

Let's look at the case where $R_{meas}$ is 2.7 ohms.  We also know that $R_{ind}$ is approximately 2.3 ohms by direct measurement.  The inductance $L$ of stainless steel is about $0.01$ mH, which is $10^{-5}$ H and the inductance of the HiperCo sample is around 10 times that for moderate applied magnetic field, so is $10^{-4}$ H.  Also using the definition $\omega = 2\pi f$.

$$ 2\pi fL \gtrsim 1/100(2.7 + 2.3) $$

$$ f \gtrsim \frac{5}{200\pi L} $$
<br>
<center>For the stainless sample, using $L \approx 10^{-5}$ H</center>

$$ f \gtrsim 800~ \textrm{Hz} $$
<br>
<center>For the HiperCo sample, using $L \approx 10^{-4}$ H</center>

$$ f \gtrsim 80~ \textrm{Hz} $$

This validates our observation about a lower limitation in being able to measure the voltage amplitude ratio with our experimental parameters.  If we want to get reliable measurements below these frequencies, we need to reduce $R_{meas}$, but due to other limitations can only reduce it to $\sim 1$ ohm.  However, the frequencies that we are able to measure are the ones most important to the experiment ($100 - 10,000$ Hz).  Therefore, we can use the measurements made with $R_{meas} = 2.7$ ohm for frequencies above $100$ Hz.

These are much more reasonable results for what we expect.  The inductance still decreases with increased frequency, but it is not dominated by a dependence on $1/f$.  To explain why there is a decrease in inductance with frequency, we will review how the sample is constructed, as well as the circuit.  The sample is a wire wrapped around the HiperCo 50 or stainless steel, both of which are metals.  We will focus on HiperCo 50 for this discussion.  We already covered the $\mu$ effect of the material, where internal magnetic domains align with applied magnetic fields, but there is a second effect from <a href="http://en.wikipedia.org/wiki/Eddy_current">Eddy currents</a>, which effectively cancel out some of the effects from the applied magnetic field.  Eddy currents are generated to oppose changes in magnetic field, which is what is happening with the applied magnetic field from the current through the coil.  Only a certain portion of the sample will have Eddy currents based on the resistivity of the HiperCo 50 material and the frequency the magnetic field is changing; this is called the <a href="http://en.wikipedia.org/wiki/Skin_effect">skin depth</a>.  The applied magnetic field will be cancelled by Eddy currents in any area within the skin depth, which effectively reduces the cross-sectional area of magnetic material, and therefore reduces the inductance.  With a large enough skin depth (at large frequencies) the total area is cancelled out.

In addition to effects on inductance, Eddy currents affect the measurements in other ways.  Previously, we have modeled the sample as an inductor in series with a DC resistor (accounting for the resistance of the wire wrappings).  The Eddy currents have an additional resistive effect because when they are induces by a changing magnetic field, the current is through resistive material, which dissipates energy to heat; same as a DC resistor.  This can be modeled with the inductor coupled as a transformer to the resistance from the Eddy currents, $R_{Eddy}$, as shown in the circuit diagram below.

<img src="figures/circuit_eddy.jpg" width="700" />

With this model, the resistance associated with the sample inductor is not a simple DC resistance; the resistance changes as a function of frequency.  This is because at higher frequencies more Eddy currents are induced with a larger skin depth, so there is more resistive dissipation in the sample.  This effective resistance can be determined by using the phase shift information in the voltage measurements between $V_{C}$ and $V_R$ as follows.

$$ V_C = (i\omega L + R_{meas} + R_{ind})I $$
$$ V_R = R_{meas}I $$

$$ A = \left|\frac{V_C}{V_R}\right| = \sqrt{\left(\frac{V_C}{V_R}\right)\left(\frac{V_C}{V_R}\right)^*} = \frac{1}{R_{meas}}\sqrt{\omega^2L^2 + (R_{meas} + R_{ind})^2} $$

We can find the phase angle of each signal by using $\theta = \textrm{tan}^{-1}(\textrm{Im}(V) / \textrm{Re}(V))$ and the phase difference $ \Delta_{\phi} = \theta_C - \theta_R$.  Since the imaginary part of $V_R = 0$, the phase shift $\theta_R = 0$, so we get:

$$ \Delta_{\phi} = \theta_C = \textrm{tan}^{-1}\left(\frac{\omega L}{R_{ind} + R_{meas}}\right) $$


$$\textrm{Solving for} \omega L:$$

$$ \omega L = R_{meas}\frac{A~ \textrm{tan}\Delta_{\phi}}{\sqrt{1 + \textrm{tan}^2 \Delta_{\phi}}} $$


$$\textrm{Solving for} L:$$

$$ L = \frac{R_{meas}}{2\pi f}\frac{A~ \textrm{tan}\Delta_{\phi}}{\sqrt{1 + \textrm{tan}^2 \Delta_{\phi}}} $$



We can also use this information to solve for $R_{ind}$, which gives us an understanding of how the sample is behaving at different frequencies.

$$ R_{ind} = R_{meas}\left(\frac{A}{\sqrt{1+\textrm{tan}^2\Delta_{\phi}}} - 1\right) $$

### Using phase model to interpret measurements

In the following analysis, the amplitude and phase measurements are used to determine the inductance and $R_{ind}$ at the measured frequencies.

```python
def determine_inductance_phase(folder, freq, R_meas):
    '''
    Used if a scope is used to measure across a resistor in series with an inductor, where the inductor also has a significant resistance. 
    The resistance with the resistor is determined using phase infomration of the signal.
    '''
    freqs = np.asarray(freq)
    V_circ = [data[folder][freq]['circuit']['amp'] for freq in freqs]
    V_r = [data[folder][freq]['resistor']['amp'] for freq in freqs]
    phase = [data[folder][freq]['circuit']['phase'] - data[folder][freq]['resistor']['phase'] for freq in freqs]
    V_circ = np.asarray(V_circ)
    V_r = np.asarray(V_r)
    phase = np.asarray(phase)
    #inductance = 1/(2*np.pi*freqs)*R_meas*np.sqrt(V_circ**2 / V_r**2 - 1)
    inductance = R_meas/(2*np.pi*freqs) * V_circ / V_r * np.tan(phase) / np.sqrt(1 + (np.tan(phase))**2)
    R_ind = R_meas * ( (V_circ / V_r) / np.sqrt(np.tan(phase) + 1) - 1 )
    
    return inductance, R_ind
```

```python
folders = [folder for folder in folders if '3ohm' in folder]
for folder in folders:
    R_meas = 2.7
    inductance, R_ind = determine_inductance_phase(folder, freqs, R_meas)
    #print inductance
    for i, freq in enumerate(freqs):
        data[folder][freq]['inductance_phase'] = inductance[i]
        data[folder][freq]['R_ind'] = R_ind[i]
```

### Determining effective $\mu$ of the material

The $\mu$ of a sample at a given $B_{external}$ is determined by the inductance measure leading back to a frequency of 0.  Because samples are expected to level out below a certain frequency, we can trace that value back to zero frequency and use that value to determine $\mu$.  For the samples that don't level out within our measurement range, we can put a minimum value on $\mu$, but would have to change the measurement conditions to measure inductance at lower frequencies to determine a $\mu$ value.

To determine $\mu$, we use the measured inductance of the sample and divide it by the calculated inductance of a sample of the same dimensions with an air core, as described in the <a href="http://hyperphysics.phy-astr.gsu.edu/hbase/electric/indsol.html#c1">HyperPhysics page</a>.  This is because an air core solenoid as a $\mu$ of 1, so we are finding the relative permeability based on that inductance value.  Using the dimensions of the sample to determine the solenoid inductance: for the HiperCo 50 $N=171$ (number of turns), $A = 0.5775$cm$^2$ (cross-sectional area), and $l = 9.5$cm (length of windings), we find the inductance to be $L = 0.0223\textrm{mH}$.

We can divide by this inductance at each frequency measurement to get an "effective $\mu$" at that frequency.  However, the meaningful $\mu$ measurement is the value that is approached decreasing towards zero frequency.  Here we see that the effective $\mu$ at $10\textrm{kGauss}$ ($1\textrm{Tesla}$) is just below 10, whereas the effective $\mu$ is around 5 at $17\textrm{kGauss}$.  We are most interested in the effective $\mu$ with a $3\textrm{kG}$ external field because that is the field inside the tokamak we work on.  This effective $\mu$ is larger than 70, which is much higher than what we expected.  We expected a value around 8.  We took a further look to figure out what was going on.

### Interpretting results with parallel $B_{external}$ compared to perpendicular

The effective $\mu$ with an external magnetic field of $3\textrm{kG}$ parallel to the applied magnetic field is around 1.1, which is significantly lower than our expected result of 8.  Because the results of a parallel field are much lower than expected and with a perpendicular field are much higher, it shows a strong dependency of the effective $\mu$ of the sample for the geometry of the $B_{applied}$ to $B_{external}$. 

### Explaining measurements of different geometries

By looking at the geometries of $B_{applied}$ compared to $B_{external}$, we can start to understand the results.  In the perpendicular case, $B_{applied}$ affects teh direction of the total magnetic field by tilting $B_{external}$.  By thinking about this at the domain level, the domains are aligned to $B_{external}$ initially, but with $B_{applied}$ adding to that field, they just need to slightly change teh direction they point back and forth to keep up with $B_{applied}$.  Whereas in the case with parallel fields, the domains are mostly pointing in the direction of $B_{external}$ and $B_{applied}$ is trying to extend and compress the magnetic field.  This is done by aligning more domains with the magnetic field or against it.  With higher $B_{external}$, more domains are already pointing in its direction, so there are less domains to be affected by $B_{applied}$.  Therefore, there is less of an inductive effect from the magnetic $\mu$ property of the material as we observed.

It has become obvious that the geometrical interpretation is very important for modeling this physics.
